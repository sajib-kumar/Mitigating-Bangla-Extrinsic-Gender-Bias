**final to run dataset**



```python
import pandas as pd
import random
import ast

def safe_convert_name_list(name_list_str):
    try:
        # Safely convert the string representation of the list to an actual list
        return ast.literal_eval(name_list_str)
    except (SyntaxError, ValueError):
        # Handle conversion errors gracefully
        return []

def add_spaces(sentence, names_list):
    for name in names_list:
        sentence = sentence.replace(name, f' {name} ')
    return sentence

def replace_names_based_on_gender(df, male_names, female_names, male_gender_term_set, female_gender_term_set):
    modified_sentences = []

    for index, row in df.iterrows():
        converted_sentence = row['Converted_sentence']
        name_list_str = row['Name']

        # Skip processing if converted_sentence is not a string (e.g., NaN or float)
        if not isinstance(converted_sentence, str):
            modified_sentences.append('')
            continue

        # Add spaces before and after certain punctuation marks
        converted_sentence = converted_sentence.replace('।', ' । ').replace(',', ' , ').replace('.', ' . ')

        name_list = safe_convert_name_list(name_list_str)


        modified_sentence = converted_sentence

        # Iterate through each name or list of names in the name_list
        for names in name_list:
            # Check if names is a list or a single string
            if isinstance(names, list):
                # Join the list of names into a single string to match against the sentence
                names_str = ' '.join(names)
            else:
                names_str = names

            # Check if the name(s) exists in the modified_sentence
            if names_str in modified_sentence:
                # Determine the gender term associated with the name(s)
                words = modified_sentence.split()

                # Check if any of the name(s) appear as a whole string in the sentence
                if any(name in words for name in names_str.split()):
                    # Check if the name(s) are followed by a female gender term
                    gender_term = None
                    for name in names_str.split():
                        name_index = words.index(name.lower()) if name.lower() in words else -1
                        if name_index != -1:
                            if name_index > 0 and words[name_index - 1].strip().lower() in female_gender_term_set:
                                gender_term = 'female'
                            elif name_index < len(words) - 1 and words[name_index + 1].strip().lower() in female_gender_term_set:
                                gender_term = 'female'
                            elif name_index > 0 and words[name_index - 1].strip().lower() in male_gender_term_set:
                                gender_term = 'male'
                            elif name_index < len(words) - 1 and words[name_index + 1].strip().lower() in male_gender_term_set:
                                gender_term = 'male'

                    if gender_term:
                        # Replace the entire names string with a single replacement name
                        if gender_term == 'female':
                            replacement_name = random.choice(female_names)
                        elif gender_term == 'male':
                            replacement_name = random.choice(male_names)

                        modified_sentence = modified_sentence.replace(names_str, replacement_name)

        modified_sentences.append(modified_sentence)

    # new DataFrame with the modified sentences
    modified_df = df.copy()
    modified_df['ner converted'] = modified_sentences  # Rename the column to 'ner converted'

    # add_spaces function
    modified_df['ner converted'] = modified_df['ner converted'].apply(lambda x: add_spaces(x, male_names + female_names))

    return modified_df


modified_df = replace_names_based_on_gender(df1, male_names, female_names, male_gender_term_set, female_gender_term_set)

# Display the modified DataFrame
print(modified_df)

# Save the result
output_csv_path = '/content/drive/MyDrive/Thesis/Thesis Restart/Final_Dataset_AferNer/Toxicity_Ner_Converted_Dataset.csv'
modified_df.to_csv(output_csv_path, index=False, encoding="utf-8-sig")

print(f"Modified DataFrame saved to: {output_csv_path}")
```
