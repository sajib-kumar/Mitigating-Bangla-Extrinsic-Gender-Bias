```python
## Gender Term Read
#gender_terms = "/content/drive/MyDrive/Thesis/GenderedTerms(Bangla)_01.csv"
df_gender_terms  = pd.read_csv("male_female.csv")
df_gender_terms.head()
```

```python
#Automatic Gender Conversion

Masculine_term = df_gender_terms['Masculine'].tolist()
Feminine_term = df_gender_terms['Feminine'].tolist()

gender_converted_df = pd.DataFrame()
original_sentences, converted_sentences = [], []

for index, row in data_df.iterrows():
  sentence = row['Comments']
  updated_sentence = sentence.split()

  final_sentences, masculine_sentences = [], []
  masculine_replacements, feminine_replacements = [], []
  masculine_idx_to_change, feminine_idx_to_change = [], []

  for word in updated_sentence:
      temp_indices = [i for i in range(len(Masculine_term)) if Masculine_term[i] == word]
      if len(temp_indices) > 0:
        masculine_replacement_indices = temp_indices
        fem_terms = [Feminine_term[x] for x in masculine_replacement_indices]
        for fem in fem_terms:
          masculine_idx_to_change.extend([i for i in range(len(updated_sentence)) if updated_sentence[i] == word])
          masculine_replacements.append(fem)

  for word in updated_sentence:
      temp_indices = [i for i in range(len(Feminine_term)) if Feminine_term[i] == word]
      if len(temp_indices) > 0:
        feminine_replacement_indices = temp_indices
        mas_terms = [Masculine_term[x] for x in feminine_replacement_indices]
        for mas in mas_terms:
          feminine_idx_to_change.extend([i for i in range(len(updated_sentence)) if updated_sentence[i] == word])
          feminine_replacements.append(mas)

  for fem_word_idx in range(len(masculine_replacements)):
    fem_word = masculine_replacements[fem_word_idx]
    new_sentence = updated_sentence.copy()
    new_sentence[masculine_idx_to_change[fem_word_idx]] = fem_word
    masculine_sentences.append(' '.join(new_sentence))


  if len(masculine_sentences) > 0:
    for mas_word_idx in range(len(feminine_replacements)):
      mas_word = feminine_replacements[mas_word_idx]
      for sentence_idx in range(len(masculine_sentences)):
        new_sentence = masculine_sentences[sentence_idx].split()
        new_sentence[feminine_idx_to_change[mas_word_idx]] = mas_word
        final_sentences.append(' '.join(new_sentence))
  else:
    for mas_word_idx in range(len(feminine_replacements)):
      mas_word = feminine_replacements[mas_word_idx]
      new_sentence = updated_sentence.copy()
      new_sentence[feminine_idx_to_change[mas_word_idx]] = mas_word
      final_sentences.append(' '.join(new_sentence))

  if len(final_sentences) == 0:
    final_sentences = masculine_sentences

  if len(final_sentences) > 0:
    original_sentences.extend([sentence]*len(final_sentences))
    converted_sentences.extend(final_sentences)

gender_converted_df['original sentence'] = original_sentences
gender_converted_df['converted sentence'] = converted_sentences
```
